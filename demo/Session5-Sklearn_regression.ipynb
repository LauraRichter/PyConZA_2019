```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

---
## Regression

#### Simulate data: Line plus gaussian noise

```python
datasize = 5000
x = np.arange(datasize)

m = 5.6
c = 5500

y = m*x + c
```

```python
fix, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(x, y)
```

```python
y = y + 800. * np.random.normal(size=datasize)
```

```python
fix, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(x, y, '.')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(x, y, '.')
ax.set_xlim([1000, 1200])
```

#### Perform linear regression

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(x, y)
```

```python
regr.fit(x.reshape(-1,1), y)
```

```python
print('Fitted m:', regr.coef_)
print('fitted c:', regr.intercept_)
```

```python
print('Simulated m:', m)
print('Simulated c:', c)
```

```python
y_predict = regr.predict(x.reshape(-1,1))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(x, y, '.')
ax.plot(x, y_predict, '-')
```

```python
# for fun let's look at the residuals
residuals = y_predict - y

fig, ax = plt.subplots(1, 1, figsize=(5,5))
ax.plot(x, residuals, '.')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,4))
ax.hist(residuals, bins=200);
```
